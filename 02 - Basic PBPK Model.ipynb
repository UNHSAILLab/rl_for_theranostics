The process is based on guidelines from the Medical Internal Radiation Dose (MIRD) pamphlets 16 and 21. Here's a breakdown:

### Data Acquisition:

1. **Planar Whole-Body Scintigraphies**: These are imaging studies that track the distribution of the radiolabeled peptides within the body. They are conducted using a double-head gamma camera.
   
2. **Regions of Interest (ROIs)**: These are manually drawn areas over source organs (e.g., tumor sites, kidneys, and parotid glands) in the imaging studies to focus the analysis. 

3. **Time of Scanning**: ROIs are initially drawn using the scan that shows the best tumor-to-background ratio, usually taken 20 or 44 hours post-injection.

### Data Processing:

1. **HERMES System**: A specific software used to analyze the scintigraphies.

2. **Quantification**: To quantify the data within each ROI, the geometric mean of anterior and posterior counts with background corrections is calculated for each organ and target lesion ROI.

3. **Time-Dependent Activities**: These are calculated based on the corrected counts for each ROI, presumably to understand how the activity changes over time in different regions.

4. **Consistency**: ROIs were always drawn by the same physicist in collaboration with a nuclear medicine physician, ensuring consistency in the data acquisition process.

5. **Choice of Lesions**: Lesions for ROI were chosen based on their high uptake and lack of overlap with other organs or lesions.

### Additional Points:

1. **CT Scans**: Used to determine the volume of the tumor lesions, parotid glands, and kidneys.
  
2. **Parotid Glands**: In this study, the term "salivary glands" is synonymous with the parotid glands. The large volume of the parotid glands makes them most suitable for activity determination from planar images.

### Implications:

1. **Precision**: The method aims for a high level of precision in tracking how the radiolabeled peptides distribute within the body, how they are taken up by target and non-target tissues, and how their activity changes over time.

2. **Safety and Effectiveness**: By closely monitoring these variables, the study aims to optimize the safety and effectiveness of the radiolabeled peptide therapy.

3. **Dosimetry**: The data will likely be used to calculate the actual radiation doses received by different tissues, which is crucial for assessing both the effectiveness of the treatment and its safety profile.






- \( \text{[RK,0]} \) (binding site density of kidneys): 27 nmol/L
- \( \text{[RSAL,0]} \) (binding site density of salivary glands): 62 nmol/L
- \( \text{[RTU,0]} \) (binding site density of tumor): 380 nmol/L
- \( \lambda_{K, \text{release}} \) (release rate of kidneys): \(2.88 \times 10^{-4} \) min\(^{-1}\)
- \( \lambda_{SAL, \text{release}} \) (release rate of salivary glands): \(3.90 \times 10^{-4} \) min\(^{-1}\)
- \( \lambda_{TU, \text{release}} \) (release rate of tumor): \(1.76 \times 10^{-4} \) min\(^{-1}\)
- \( f_{SAL} \) (serum flow per mass of salivary glands): \(0.20 \, \text{ml} \, \text{g}^{-1} \, \text{min}^{-1} \)


```python
import numpy as np
from matplotlib import pyplot as plt
# Re-executing the Python code for the extended PBPK model including Salivary Glands with new parameters
# Time step (minutes)
dt = 0.1

# Time array
time = np.arange(0, 100, dt)  # 0 to 100 minutes

# Initialize arrays for concentrations
blood = np.zeros_like(time)
tumor = np.zeros_like(time)
kidney = np.zeros_like(time)
salivary_gland = np.zeros_like(time)

# Initial concentration in the blood
blood[0] = 100  # Arbitrary units

# Parameters from tables
# For simplicity, all rates are converted to arbitrary units compatible with the time step
extravasation_rate_tumor = 0.1  # Placeholder
extravasation_rate_kidney = 0.05  # Placeholder
extravasation_rate_salivary = 0.03  # Placeholder
binding_rate_tumor = 380  # [RTU,0] in nmol/L
binding_rate_kidney = 27  # [RK,0] in nmol/L
binding_rate_salivary = 62  # [RSAL,0] in nmol/L
release_rate_tumor = 1.76e-4  # λTU,release in min-1
release_rate_kidney = 2.88e-4  # λK,release in min-1
release_rate_salivary = 3.90e-4  # λSAL,release in min-1
serum_flow_salivary = 0.20  # fSAL in ml g-1 min-1

# Other parameters remain the same
internalization_rate = 0.005
degradation_rate = 0.002
decay_rate = 0.0001
clearance_rate = 0.01

# Simulation loop
for i in range(1, len(time)):
   # Update concentrations based on new parameters
   blood_to_tumor = extravasation_rate_tumor * blood[i-1] * dt
   blood_to_kidney = extravasation_rate_kidney * blood[i-1] * dt
   blood_to_salivary = extravasation_rate_salivary * blood[i-1] * dt

   tumor_to_blood = release_rate_tumor * tumor[i-1] * dt
   kidney_to_blood = release_rate_kidney * kidney[i-1] * dt
   salivary_to_blood = release_rate_salivary * salivary_gland[i-1] * dt

   binding_tumor = binding_rate_tumor * tumor[i-1] * dt
   binding_kidney = binding_rate_kidney * kidney[i-1] * dt
   binding_salivary = binding_rate_salivary * salivary_gland[i-1] * dt

   internalization_tumor = internalization_rate * binding_tumor * dt
   internalization_kidney = internalization_rate * binding_kidney * dt
   internalization_salivary = internalization_rate * binding_salivary * dt

   degradation_tumor = degradation_rate * internalization_tumor * dt
   degradation_kidney = degradation_rate * internalization_kidney * dt
   degradation_salivary = degradation_rate * internalization_salivary * dt

   decay = decay_rate * (blood[i-1] + tumor[i-1] + kidney[i-1] + salivary_gland[i-1]) * dt
   clearance = clearance_rate * blood[i-1] * dt

   # Update blood concentration
   blood[i] = blood[i-1] - (blood_to_tumor + blood_to_kidney + blood_to_salivary + clearance + decay)

   # Update tumor concentration
   tumor[i] = tumor[i-1] + blood_to_tumor - (tumor_to_blood + binding_tumor + internalization_tumor + degradation_tumor + decay)

   # Update kidney concentration
   kidney[i] = kidney[i-1] + blood_to_kidney - (kidney_to_blood + binding_kidney + internalization_kidney + degradation_kidney + decay)

   # Update salivary gland concentration
   salivary_gland[i] = salivary_gland[i-1] + blood_to_salivary - (salivary_to_blood + binding_salivary + internalization_salivary + degradation_salivary + decay)

# Plotting the results
plt.plot(time, blood, label='Blood')
plt.plot(time, tumor, label='Tumor')
plt.plot(time, kidney, label='Kidney')
plt.plot(time, salivary_gland, label='Salivary Gland')
plt.xlabel('Time (minutes)')
plt.ylabel('Concentration (arbitrary units)')
plt.legend()
plt.title('Extended PBPK Model with New Parameters')
plt.show()
```
